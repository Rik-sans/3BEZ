```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#df = pd.read_csv('titanic.csv')

df = sns.load_dataset('titanic')
age_survival = df.groupby('age')['survived'].mean()
plt.figure(figsize=(12, 6))
plt.plot(age_survival.index, age_survival.values, color='blue', linewidth=2)
plt.title('     (Titanic Dataset)', fontsize=14)
plt.xlabel('', fontsize=12)
plt.ylabel(' ', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

#display(df.head())

#display(df.info())

#display(df.describe())

#women = df.query("Survived == 1")

#sns.lineplot(data=women, x = "age", y=)

```
